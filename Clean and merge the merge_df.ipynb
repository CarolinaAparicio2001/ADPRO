Clean and merge the merge_df

```python
import pandas as pd
```

```python
airlines= pd.read_csv('airlines.csv')
airplanes= pd.read_csv('airplanes.csv')
airports = pd.read_csv('airports.csv')
routes =  pd.read_csv('routes.csv')
```

```python
# Drop the first column and reset the index
airlines_df = airlines.drop(airlines.columns[0], axis=1).reset_index(drop=True)
airports_df= airports.drop(airports.columns[0], axis=1).reset_index(drop=True)
airports_df = airports_df.drop(['Type', 'Source'], axis=1)
routes_df= routes.drop(routes.columns[0], axis=1).reset_index(drop=True)
airplanes_df= airplanes.drop(airplanes.columns[0], axis=1).reset_index(drop=True)
```

```python
merge_df_1 = pd.merge(routes_df, airports_df, left_on="Source airport", right_on="IATA")
merge_df_1 = merge_df_1.rename(columns={"Country": "Source country","Latitude": "latitude_source","Longitude": "longitude_source"})
merge_df_1.head(5)
```

```python
# Assuming merge_df is your DataFrame
column_names = merge_df_1.columns.tolist()

# Print or use the column names as needed
print(column_names)
```

```python
merge_df_2 = pd.merge(merge_df_1,airports_df[["IATA", "Country", "Latitude", "Longitude"]],left_on="Source airport",right_on="IATA")
merge_df_2 = merge_df_2.rename(columns={'IATA_x':"IATA", "Country": "Destination country","Latitude": "latitude_destination","Longitude": "longitude_destination"})
merge_df_2
```

```python
# Assuming merge_df is your DataFrame
column_names = merge_df_2.columns.tolist()

# Print or use the column names as needed
print(column_names)
```

```python
merge_df_2 = merge_df_2.dropna(subset=["Source country", "Destination country"])
merge_df_2
```

```python
#Drop the columns that I dont need
merge_df_2.drop(columns=['IATA_y'], inplace=True)
```

```python
# Assuming merge_df is your DataFrame
column_names = merge_df_2.columns.tolist()

# Print or use the column names as needed
print(column_names)
```

```python
merge_df_3=pd.merge(merge_df_2,airlines_df,left_on="Airline",right_on="IATA")
merge_df_3 = merge_df_3.rename(columns={'Name_y':"Name Airline",'IATA_x':'IATA', 'ICAO_x':'ICAO','Name_x':'Airport Name','Airline ID_x':'Airline ID'})
merge_df_3.drop(columns=['Airline ID_y', 'IATA_y','ICAO_y','Alias','Airport ID','Timezone', 'DST', 'Tz database time zone'], inplace=True)
merge_df_3
```

```python
# Assuming merge_df is your DataFrame
column_names = merge_df_3.columns.tolist()

# Print or use the column names as needed
print(column_names)
```

```python
airplanes_df = airplanes_df.rename(columns={'IATA code':"IATA"})
merge_df=pd.merge(merge_df_3,airplanes_df, how="left")
merge_df
```
