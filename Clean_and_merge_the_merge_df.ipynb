```python
import pandas as pd
```

```python
airlines= pd.read_csv('airlines.csv')
airplanes= pd.read_csv('airplanes.csv')
airports = pd.read_csv('airports.csv')
routes =  pd.read_csv('routes.csv')
```

```python
# Drop the first column and reset the index
airlines_df = airlines.drop(airlines.columns[0], axis=1).reset_index(drop=True)
airports_df= airports.drop(airports.columns[0], axis=1).reset_index(drop=True)
airports_df = airports_df.drop(['Type', 'Source'], axis=1)
routes_df= routes.drop(routes.columns[0], axis=1).reset_index(drop=True)
airplanes_df= airplanes.drop(airplanes.columns[0], axis=1).reset_index(drop=True)
```

```python
merge_df_1 = pd.merge(airports_df,routes_df, left_on="IATA", right_on="Source airport", how="left")
merge_df_1 = merge_df_1.rename(columns={"Country": "Source country","Latitude": "latitude_source","Longitude": "longitude_source"})
merge_df_1
```
